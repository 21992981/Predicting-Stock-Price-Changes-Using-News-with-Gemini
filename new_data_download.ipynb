```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
df = pd.read_csv("stock_data_daily.csv")
df.Date = pd.to_datetime(df.Date)
df = df.set_index('Date')
```

```python
df.tail()
```

```python
last_day = df.tail(1)
last_day_index = last_day.index
```

```python
last_day
```

```python
last_day_index
```

```python
start_str = str(last_day_index[0]).split(' ')[0]
```

```python
def add_stock (index, main_df):
    new_df = yf.download(index, start=start_str, interval="1d")
    stock_score = 'score'
    st_sc = stock_score + "_" + index
    new_df[stock_score] = float(df.loc[last_day_index, st_sc].iloc[0])
    new_df = new_df.add_suffix('_' + index)
    main_df = main_df.join(new_df,how='left')
    return main_df
```

```python
new_val = yf.download('XU100.IS', start=start_str, interval="1d")
new_val = new_val.add_suffix('_' + 'XU100')
with open("bist_100.txt") as f:
    bist_list = f.read().splitlines()

for line in bist_list:
    index = line + ".IS"
    new_val = add_stock(index, new_val)
```

```python
df = df[:-1]
```

```python
result = pd.concat([df, new_val], axis=0)
result
```

```python
result.to_csv('stock_data_daily.csv')
```
