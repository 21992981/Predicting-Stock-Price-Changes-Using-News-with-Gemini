```python
import yfinance as yf
```

```python
start_str = '2024-01-01'
```

```python
def add_stock (index, main_df):
    new_df = yf.download(index, start=start_str, interval="1d")
    stock_score = 'score'
    new_df[stock_score] = 50
    new_df = new_df.add_suffix('_' + index)
    main_df = main_df.join(new_df,how='left')
    return main_df
```

```python
df = yf.download('XU100.IS', start=start_str, interval="1d")
df = df.add_suffix('_' + 'XU100')
with open("bist_100.txt") as f:
    bist_list = f.read().splitlines()

for line in bist_list:
    index = line + ".IS"
    df = add_stock(index, df)
```
